```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Data/student_depression_dataset.csv')
```

```python
df.pop('id')
```

```python
new_df = df.head(50)
```

```python
new_df
```

```python
new_df.plot(x = 'Financial Stress', y = 'Age', kind='bar')
plt.show()
```

```python
new_df.plot(x='Financial Stress', y='Age', kind='scatter')
plt.show()
```

```python
new_df.plot.pie(y='Age', figsize=(5, 5))
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
print(df.head())
```

```python
sns.kdeplot(df['sepal_length'], fill=True)
```

```python
for col in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    sns.kdeplot(df[col], fill=True)
```

```python
sns.pairplot(df, hue='species')
```
