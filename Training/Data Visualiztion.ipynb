```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Data/student_depression_dataset.csv')
```

```python
df.pop('id')
```

```python
new_df = df.head(10)
```

```python
new_df
```

```python
new_df.plot(x = 'Financial Stress', y = 'Age', kind='bar')
plt.show()
```

```python
new_df.plot(x='Financial Stress', y='Age', kind='scatter')
plt.show()
```

```python
new_df.plot.pie(y='Age', figsize=(5, 5))
```
