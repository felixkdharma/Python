```python
""" Lists = Values in lists are called as elements or items 
    Mutable
    Variable Length
    Accessed by Indexing and Slicing
"""
""" Tuples = similar to lists but the diffrence is ended with parentheses () 
    Immutable
    Variable Length
    Accessed by Indexing and Slicing
"""
```

```python
cust_info = ["CUST001", "John Doe", 28, ["UK", "12 Des 1999"]]
cust_info2  = ("CUST002", "Jane Smith", 32)

dob_cust_info = cust_info[3]
print(dob_cust_info) 
```

```python
## Indexing Tuples
cust_info2[1]
## Slicing Tuples
cust_info2[1:3]
```

```python
## Tuple Concattenation

cust_info3 = ("CUST003", "Alice Johnson", 25)
all_cust = cust_info2 + cust_info3
print(all_cust)

cust_info4 = cust_info2 + cust_info3 + ("CUST004", "Bob Brown", 30)
print(cust_info4)
```

```python
## Min and Max Functions will not work with tuples having mixed data types
exmp_tuple = ("a", "vb")
max(exmp_tuple) ## will be 'vb' as it is lexicographically higher than 'a'

exmp_tupl2 = (1,100,110)
max(exmp_tupl2) ## will be 110
```

```python
## Immutability
## In fact tuples cannot directly be changed the value of elements we can slice it and store to the new value

exp_tuple = ("Hello", "My Name is", 12, 13, "Felix Kurniawan Dharma")
var_slice1 = exp_tuple[0:2]
var_slice2 = exp_tuple[4]

# new_tuple = var_slice1 + var_slice2
new_tuple = exp_tuple[0:2] + (var_slice2,)
print(new_tuple)
```

```python
# Sorting Tuples

tuple1 = (5,3,8,1,2)
sorted_tuple = tuple(sorted(tuple1))
print(sorted_tuple)
```
