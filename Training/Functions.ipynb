```python
## Functions

def even_or_odd():
    input_number = int(input("Enter a number: "))
    if (input_number % 2) == 0:
        print(f"{input_number} is even.")
    else:
        print(f"{input_number} is odd.")
```

```python
even_or_odd()
```

```python
## Factorial Function
## Example 4! = 4 x 3 x 2 x 1 = 24

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
factorial(6)
```

```python
## Lambda Functions

square = lambda x: x * x
square(5)
```

```python
even_odd = lambda x: "even" if x % 2 == 0 else "odd"
print("Number", 10, even_odd(10))
```
