Numpy

can be used to store data of diffrent forms: 1-d, 2-d, 3-d

can be used to computation and operation single and multi-dimensional array of elements

np.ndim => for checking dimensional array

np.array => creating a x dimensional array

np.len => get a length of array

np.shape => get how many rows

np[x,y] => to get an index of array elements

np[:, x] => will get a specific columns

np.zeros((x,y)) => will create a x dimensional array with 0 value

np.ones((x,y)) => will create a x dimensional array with 1 value

np.full((x,y), z, dtype) => will create a x dimensional array with constant variable 

np.random.rand((x,y)) => will create a matrix with a random value (dtype = float)

np.random.randint(range, ,size(x,y) )=> will create a matrix with a random value (dtype=int) and have range value

```python
import numpy as np
```

```python
## numpy 1d array
arr = np.array([1, 2, 3, 4, 5])
print("1D Array:", arr)

## numpy 2d array
arr2 = np.array([[1,2,3], [3,2,1]], dtype=np.int32)
print("2D Array:\n", arr2)

print("arr has : ", arr.shape)
print("arr2 has: ", arr2.shape)
```
