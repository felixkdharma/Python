```python
## File Handling ##

# Basic functions: open(), read(), close()
# obj = open(file_name[,access_mode][,buffering])

#r means read only
#w means write only (overwrites existing file or creates new file)
```

```python
file_txt = open('test.txt', 'r')

print(file_txt.read())
file_txt.close()
```

```python
file_txt = open('../Data/test.txt', 'w')
file_txt.write("Felix Kurniawan Dharma")
file_txt.close()
```

```python
file_txt = open("../Data/test.txt", "r")
content = file_txt.read()
print(content)
```

```python
## Append Operation
file_txt = open("../Data/test.txt", "a")
file_txt.write("\nAdditional line added.")
print("File read successfully.")
file_txt.close()
```
