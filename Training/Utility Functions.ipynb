```python
 ## Map, Reduce, and Filter Functions

# Map takes 2 arguments first is a function and second is an iterable
# map(function, iterable)

import math

# Area of a circle
radii = [14,21,28]
area = []

## Default Functions
def circle_area(r):
    return math.pi * (r**2)

for i in radii:
    area.append(circle_area(i))

## Map Function with lambda function
area = list(map(lambda x: math.pi * (x**2), radii))
area
```

```python
# Convert Celcius scale to Farheneit

temps = [("Berlin",29),("Cairo",36),("Buenos Aires",19),("Los Angeles",26),("Tokyo",27),("New York",28),("Lagos",33)]
## Using map with lambda function
## [0] and [1] are used to access the key values in the dictionary

cel_to_far = list(map(lambda data: (data[0], (9/5)*data[1] + 32), temps))
cel_to_far
```

```python
# Filter Function

import statistics

# .mean() function to calculate the mean of a list
data = [1,3,5,7,8,12,11,13]

avg = statistics.mean(data)
print(avg)

# filter list that value greater than the mean value
list(filter(lambda x: x > avg, data))

#filtering values that contain default value

data2 = ["felix", 0, 0.0, "kurniawan", ""]
list(filter(None, data2))
```
