```python
## Defining A Class

class rectangle:
    def __init__(self):
        self.length = 5
        self.breadth = 10
```

```python
rect = rectangle()
print("Length:", rect.length, "\nBreadth:", rect.breadth)
```

```python
## Parameterized Constructor
class rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth   
```

```python
rect2 = rectangle(15, 25)
print("Length:", rect2.length, "\nBreadth:", rect2.breadth)

rect2.area()
```

```python
##Inheritance and Overriding
# in python the code will be using a pass. and will inherit all variables and methods from the parent class

class Employee:        
    def function1(self):
        print("Employee Created")

class Department(Employee):
    pass
```

```python
felix = Department()
felix.function1()
```
