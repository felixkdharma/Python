```python
### Iterate over a Dictionary
student_data = {1: ["Felix", 22],
                2: ["Alice", 23],
                3: ["Bob", 24]}
for key, value in student_data.items():
    print("Student ID:", key)
    print("Name:", value[0])
    print("Age:", value[1])
```
